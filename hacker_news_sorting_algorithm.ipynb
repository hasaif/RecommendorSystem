```python
import numpy as np  
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import sem, t

ratings_data = pd.read_csv("ml-latest-small\\ratings.csv")  
movie_names = pd.read_csv("ml-latest-small\\movies.csv")  
movie_data = pd.merge(ratings_data, movie_names, on='movieId')  
movie_data.head()
```

```python
#https://medium.com/hacking-and-gonzo/how-hacker-news-ranking-algorithm-works-1d9b0cf2c08d
def score(rating, time):
     return (rating-1)/ (time+2)**1.8


hackern = pd.DataFrame(movie_data.groupby('title')['rating'].sum()) 
hackern['time']= pd.DataFrame(movie_data.groupby('title')['timestamp'].min())
hackern['score'] = np.vectorize(score)(hackern['rating'],hackern['time'])


hackern.sort_values(by=['score']).tail(10)


```
